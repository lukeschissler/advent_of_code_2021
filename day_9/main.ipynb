Part 1

```python
ins = open('input.txt', 'r')
ins = ins.read().splitlines()
ins = [list(x) for x in ins]
```

```python
def get_safe_indices(x, y, max_x, max_y):
    outs = []
    if x != 0:
        outs.append([x-1, y])
    if y != 0:
        outs.append([x, y-1])
    if x != max_x:
        outs.append([x+1, y])
    if y != max_y:
        outs.append([x, y+1])
        
    return outs
```

```python
def part_1(ins):
    outs = []
    eliminated = []
    max_y = len(ins)-1
    max_x = len(ins[0])-1

    for y, row in enumerate(ins):
        for x, curr_val in enumerate(row):
            if [y, x] in eliminated:
                continue

            safe_indices = get_safe_indices(x, y, max_x, max_y)
            local_minimum = []

            for check_y, check_x in safe_indices:
                if int(ins[check_y][check_x]) > int(curr_val):
                    local_minimum.append(True)
                    eliminated.append([check_y, check_x])

            if len(local_minimum) == len(safe_indices):
                outs.append(int(curr_val))

    return outs
```

```python
outs = part_1(ins)
outs = sum([x + 1 for x in outs])
outs
```

Part 2

```python
def part_2(ins):
    seen_indices = []
    basin_sizes = []
    max_y = len(ins)-1
    max_x = len(ins[0])-1

    def scan_basin(x, y):
        if int(ins[y][x]) != 9:
            seen_indices.append([x, y])
            safe_indices = get_safe_indices(x, y, max_x, max_y)
            safe_indices = [x for x in safe_indices if x not in seen_indices]        
            return 1 + sum([scan_basin(x, y) for x, y in safe_indices])
        else: 
            return 0

    for y, row in enumerate(ins):
        for x, curr_val in enumerate(row):
            if [x, y] in seen_indices:
                continue
            curr_basin = scan_basin(x, y)
            basin_sizes.append(curr_basin)
    return basin_sizes
```

```python
basins = part_2(ins)
basins.sort(reverse=True)
basins
```
